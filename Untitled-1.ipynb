```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import warnings
warnings.filterwarnings('ignore')

# 1. Load & clean
df = pd.read_csv('85334965-5736-457a-b8d4-a077e6872f84.csv')
df_clean = df.dropna()

# 2. EDA
print(df_clean.corr()['Sales'].sort_values(ascending=False))

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
for ax, col, color in zip(axes.flatten()[:3],
                           ['TV', 'Radio', 'Social_Media'],
                           ['steelblue', 'darkorange', 'green']):
    ax.scatter(df_clean[col], df_clean['Sales'], alpha=0.3, color=color, s=10)
    r = df_clean[col].corr(df_clean['Sales'])
    ax.set_title(f'{col} vs Sales  (r = {r:.3f})')
sns.heatmap(df_clean.corr(), annot=True, fmt='.3f', cmap='coolwarm', ax=axes[1,1])
plt.tight_layout(); plt.show()

# 3. OLS Regression
X_const = sm.add_constant(df_clean['TV'])
model   = sm.OLS(df_clean['Sales'], X_const).fit()
print(model.summary())

# 4. Diagnostics
fitted    = model.fittedvalues
residuals = model.resid
std_resid = residuals / residuals.std()

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes[0,0].scatter(fitted, residuals, alpha=0.3, s=10); axes[0,0].axhline(0, color='red', ls='--'); axes[0,0].set_title('Residuals vs Fitted')
sm.qqplot(residuals, line='s', ax=axes[0,1], alpha=0.4); axes[0,1].set_title('Q-Q Plot')
axes[1,0].scatter(fitted, np.sqrt(np.abs(std_resid)), alpha=0.3, s=10, color='green'); axes[1,0].set_title('Scale-Location')
x_r = np.linspace(df_clean['TV'].min(), df_clean['TV'].max(), 200)
axes[1,1].scatter(df_clean['TV'], df_clean['Sales'], alpha=0.3, s=10)
axes[1,1].plot(x_r, model.params['const'] + model.params['TV']*x_r, color='red', lw=2)
axes[1,1].set_title('Regression Line')
plt.tight_layout(); plt.show()

# 5. Report
print(f"Sales = {model.params['const']:.2f} + {model.params['TV']:.2f} × TV")
print(f"R² = {model.rsquared:.3f}")
```
